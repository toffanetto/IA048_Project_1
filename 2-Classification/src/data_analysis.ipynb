```python
# @toffanetto
import numpy as np
import logistic_regression_classification as cl
import matplotlib.pyplot as plt

# Getting train data
X_train, y_train, classes_rate_train = cl.getData(train=True, raw= False)

# Getting test data
X_test, y_test, classes_rate_test = cl.getData(train=False, raw= False)
```

```python
barWidth = 0.4

plt.figure()
 
# Set position of bar on X axis 
br1 = np.arange(len(classes_rate_train)) 
br2 = [x + barWidth for x in br1] 
 
# Make the plot
plt.bar(br1, classes_rate_train, color ='b', width = barWidth, 
        edgecolor ='grey', label ='Training data') 
plt.bar(br2, classes_rate_test, color ='g', width = barWidth, 
        edgecolor ='grey', label ='Test data') 
 
# Adding Xticks 
plt.xlabel('Class') 
plt.ylabel('Rate') 
plt.xticks([r + barWidth/2 for r in range(len(classes_rate_train))], 
        ['1', '2', '3', '4', '5', '6'])
plt.title('Balancing of classes in dataset')
 
plt.legend()

plt.savefig("../plot/Balancing_of_classes.pdf", format="pdf", bbox_inches="tight")
```
